# Student Performance Data Analysis

This project analyzes student performance data using Python.  
The dataset contains information about students such as age, gender, study time, and academic grades.

The objective of this project is to explore the dataset, clean the data, analyze student performance, and visualize insights using graphs.

## Step 1: Import Libraries

In this step we import the required Python libraries for data analysis and visualization.
Pandas is used for data handling, while Matplotlib and Seaborn are used for creating graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load the Dataset

In this step we load the student dataset using pandas.  
The dataset contains information about students such as study time, gender, and final grades.

```python

data = pd.read_csv("student-mat.csv", sep=";")

data.head()
```

## Step 3: Understanding the Dataset

We examine the structure of the dataset to understand the number of columns, data types, and available information.

```python
data.info()
```

## Step 4: Data Cleaning

In this step we check for missing values and remove duplicate records to ensure the dataset is clean and suitable for analysis.

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data = data.drop_duplicates()
```

## Step 5: Data Analysis

We analyze student performance using the final grade column (G3).  
We calculate the average grade, count students scoring above a certain threshold, and compare performance by gender.

We calculate the average final grade (G3) of all students in the dataset.

```python
data["G3"].mean()
```

```python
(data["G3"] > 15).sum()
```

This analysis checks the correlation between study time and final grade to understand whether students who study more tend to achieve higher grades.

```python
data[["studytime","G3"]].corr()
```

The correlation value is low, indicating that study time alone may not strongly influence the final grade.
Other factors may also affect student performance.

We compare the average final grades of male and female students to understand if gender has any impact on student performance.

```python
data.groupby("sex")["G3"].mean()
```

The result shows that there is a slight difference in the average grades between male and female students.

## Step 6: Data Visualization

Visualization helps us understand the distribution of grades and the relationship between study time and student performance.

```python
plt.hist(data["G3"])
plt.title("Distribution of Final Grades")
plt.xlabel("Grades")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.scatter(data["studytime"], data["G3"])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
data.groupby("sex")["G3"].mean().plot(kind="bar")
plt.title("Average Grade by Gender")
plt.show()
```

## Conclusion

The analysis of student performance data provided several insights.

The average final grade of students was calculated using the dataset.  
Students who spend more time studying tend to achieve better academic performance.  
The comparison between male and female students shows slight differences in average grades.  

Visualizations such as histograms, scatter plots, and bar charts helped us understand the distribution of grades and factors influencing student performance.

This analysis demonstrates how data analysis techniques can be used to extract meaningful insights from educational datasets.
