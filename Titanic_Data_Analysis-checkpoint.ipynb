## 📂 Section 2 – Data Loading

We load the Titanic dataset from a CSV file.

```python
# Option 1: Load from CSV file (Kaggle download)
# df = pd.read_csv('titanic.csv')

# Option 2: Load directly from Seaborn's built-in dataset (no download needed)
df = sns.load_dataset('titanic')

print(f"✅ Dataset loaded successfully!")
print(f"📊 Shape: {df.shape[0]} rows × {df.shape[1]} columns")
```

```python
# Preview the first 5 rows
print("🔍 First 5 rows of the dataset:")
df.head()
```

```python
# Get basic information about the dataset
print("📋 Dataset Info:")
df.info()
```

```python
# Statistical summary of numerical columns
print("📈 Statistical Summary:")
df.describe()
```

```python
# Check for missing values
missing = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing,
    'Percentage (%)': missing_percent.round(2)
})

print("❌ Missing Values Summary:")
print(missing_df[missing_df['Missing Values'] > 0])
```

### 💡 Insight – Data Loading
- The dataset has **891 passengers** and multiple features.
- **`age`** has the most missing values (~20%), which we will handle in the cleaning step.
- **`deck`/`cabin`** has too many missing values (>77%) — it will be dropped.
- **`embarked`** is missing only 2 values — easy to fix.

---
## 🧹 Section 3 – Data Cleaning

Real-world data is rarely perfect. We need to:
1. Handle **missing values** in `age`, `embarked`
2. **Drop columns** that are not useful for analysis
3. Verify **data types** are correct

```python
# Step 1: Fill missing 'age' values with the median age
# Median is preferred over mean because it is less affected by outliers

median_age = df['age'].median()
print(f"📌 Median Age: {median_age}")

df['age'].fillna(median_age, inplace=True)

print(f"✅ Missing 'age' values filled with median: {median_age}")
print(f"   Remaining missing in 'age': {df['age'].isnull().sum()}")
```

```python
# Step 2: Fill missing 'embarked' with the most common port (mode)
most_common_port = df['embarked'].mode()[0]
df['embarked'].fillna(most_common_port, inplace=True)

print(f"✅ Missing 'embarked' filled with mode: '{most_common_port}'")
print(f"   Remaining missing in 'embarked': {df['embarked'].isnull().sum()}")
```

```python
# Step 3: Drop columns with too many missing values or not useful for analysis
# 'deck' has ~77% missing, 'embark_town' and 'who' are duplicates of other info

cols_to_drop = ['deck', 'embark_town', 'alive', 'alone', 'adult_male']

# Only drop columns that actually exist in the dataframe
cols_to_drop = [col for col in cols_to_drop if col in df.columns]
df.drop(columns=cols_to_drop, inplace=True)

print(f"✅ Dropped columns: {cols_to_drop}")
print(f"📊 New shape: {df.shape}")
```

```python
# Step 4: Verify data types
print("📋 Data Types after Cleaning:")
print(df.dtypes)
print()

# Ensure 'survived' and 'pclass' are correct types
df['survived'] = df['survived'].astype(int)
df['pclass'] = df['pclass'].astype(int)

print("✅ Data types verified!")
```

```python
# Final check for missing values after cleaning
print("🔍 Missing Values After Cleaning:")
print(df.isnull().sum())
print()
print(f"✅ Dataset is clean! Shape: {df.shape}")
```

### 💡 Insight – Data Cleaning
- **Age** missing values were filled with the **median age (28 years)** — a safe and unbiased approach.
- **Embarked** was filled with the **most common port 'S' (Southampton)**.
- Columns with too many missing values or redundant information were **dropped** to keep the analysis clean.
- The dataset is now **ready for analysis**!

## 🔍 Section 4 – Data Analysis

Now we will answer three key business questions:
1. Who survived more — **males or females**?
2. Did **passenger class** affect survival chances?
3. What was the **survival rate by age group**?

### ❓ Question 1: Who Survived More — Males or Females?

```python
# Count of survivors by gender
gender_survival_count = df.groupby('sex')['survived'].sum()

# Survival rate by gender
gender_survival_rate = df.groupby('sex')['survived'].mean() * 100

gender_summary = pd.DataFrame({
    'Total Passengers': df['sex'].value_counts(),
    'Survivors': gender_survival_count,
    'Survival Rate (%)': gender_survival_rate.round(2)
})

print("👥 Survival Summary by Gender:")
print(gender_summary)
```

#### 💡 Insight – Gender
- **Females had a much higher survival rate (~74%)** compared to males (~19%).
- This reflects the **"women and children first"** evacuation policy followed during the disaster.
- Although there were more male passengers, far fewer survived.

### ❓ Question 2: Did Passenger Class Affect Survival Chances?

```python
# Count of survivors by passenger class
class_survival_count = df.groupby('pclass')['survived'].sum()

# Survival rate by passenger class
class_survival_rate = df.groupby('pclass')['survived'].mean() * 100

class_summary = pd.DataFrame({
    'Total Passengers': df['pclass'].value_counts().sort_index(),
    'Survivors': class_survival_count,
    'Survival Rate (%)': class_survival_rate.round(2)
})

class_summary.index = ['1st Class', '2nd Class', '3rd Class']

print("🎫 Survival Summary by Passenger Class:")
print(class_summary)
```

#### 💡 Insight – Passenger Class
- **1st class passengers had the highest survival rate (~63%)**, likely because they had easier access to lifeboats.
- **2nd class had a moderate rate (~47%)**.
- **3rd class had the lowest survival rate (~24%)**, as they were located deeper in the ship and had limited lifeboat access.
- This reveals a clear **socioeconomic disparity** in survival chances.

### ❓ Question 3: What Was the Survival Rate by Age Group?

```python
# Create age groups using pd.cut()
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child (0-12)', 'Teen (13-18)', 'Young Adult (19-35)',
          'Middle-Aged (36-60)', 'Senior (60+)']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=True)

# Survival rate by age group
age_survival_count = df.groupby('age_group', observed=True)['survived'].sum()
age_survival_rate = df.groupby('age_group', observed=True)['survived'].mean() * 100
age_total = df['age_group'].value_counts().sort_index()

age_summary = pd.DataFrame({
    'Total Passengers': age_total,
    'Survivors': age_survival_count,
    'Survival Rate (%)': age_survival_rate.round(2)
})

print("🎂 Survival Summary by Age Group:")
print(age_summary)
```

#### 💡 Insight – Age Group
- **Children (0–12) had the highest survival rate**, consistent with the priority given to children during evacuation.
- **Young adults and middle-aged passengers** had moderate survival rates.
- **Seniors (60+) had a lower survival rate**, possibly due to physical limitations during evacuation.
- The age group analysis reinforces the **"children first"** evacuation policy.

## 📊 Section 5 – Data Visualization

Visualizations help us communicate findings clearly and quickly.  
We will create:
1. **Bar chart** – Survival by Gender
2. **Bar chart** – Survival by Passenger Class
3. **Histogram** – Age Distribution of Passengers
4. **Bar chart** – Survival Rate by Age Group *(bonus)*

```python
# ── Plot 1: Survival Rate by Gender ──────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Count of survivors
survival_gender_count = df.groupby(['sex', 'survived']).size().unstack()
survival_gender_count.columns = ['Did Not Survive', 'Survived']
survival_gender_count.plot(kind='bar', ax=axes[0], color=['#e74c3c', '#2ecc71'],
                            edgecolor='black', width=0.6)
axes[0].set_title('Survival Count by Gender', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Gender', fontsize=12)
axes[0].set_ylabel('Number of Passengers', fontsize=12)
axes[0].set_xticklabels(['Female', 'Male'], rotation=0)
axes[0].legend(fontsize=10)

# Add value labels on bars
for container in axes[0].containers:
    axes[0].bar_label(container, fontsize=10)

# Right: Survival rate
gender_rate = df.groupby('sex')['survived'].mean() * 100
colors = ['#e91e8c', '#1e90ff']
bars = axes[1].bar(gender_rate.index, gender_rate.values,
                   color=colors, edgecolor='black', width=0.4)
axes[1].set_title('Survival Rate (%) by Gender', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Gender', fontsize=12)
axes[1].set_ylabel('Survival Rate (%)', fontsize=12)
axes[1].set_ylim(0, 100)
axes[1].set_xticklabels(['Female', 'Male'], fontsize=11)

# Add percentage labels on bars
for bar in bars:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height + 1,
                 f'{height:.1f}%', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.suptitle('Figure 1: Gender vs Survival', y=1.02, fontsize=15, fontweight='bold')
plt.savefig('plot1_gender_survival.png', bbox_inches='tight', dpi=150)
plt.show()
print("✅ Plot 1 saved as 'plot1_gender_survival.png'")
```

```python
# ── Plot 2: Survival by Passenger Class ──────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Count of survivors by class
class_count = df.groupby(['pclass', 'survived']).size().unstack()
class_count.columns = ['Did Not Survive', 'Survived']
class_count.index = ['1st Class', '2nd Class', '3rd Class']
class_count.plot(kind='bar', ax=axes[0], color=['#e74c3c', '#2ecc71'],
                  edgecolor='black', width=0.6)
axes[0].set_title('Survival Count by Passenger Class', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Passenger Class', fontsize=12)
axes[0].set_ylabel('Number of Passengers', fontsize=12)
axes[0].set_xticklabels(['1st Class', '2nd Class', '3rd Class'], rotation=0)
axes[0].legend(fontsize=10)

for container in axes[0].containers:
    axes[0].bar_label(container, fontsize=10)

# Right: Survival rate by class
class_rate = df.groupby('pclass')['survived'].mean() * 100
bar_colors = ['#f39c12', '#3498db', '#9b59b6']
bars = axes[1].bar(['1st Class', '2nd Class', '3rd Class'],
                    class_rate.values, color=bar_colors, edgecolor='black', width=0.5)
axes[1].set_title('Survival Rate (%) by Passenger Class', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Passenger Class', fontsize=12)
axes[1].set_ylabel('Survival Rate (%)', fontsize=12)
axes[1].set_ylim(0, 100)

for bar in bars:
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width()/2., height + 1,
                 f'{height:.1f}%', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.suptitle('Figure 2: Passenger Class vs Survival', y=1.02, fontsize=15, fontweight='bold')
plt.savefig('plot2_class_survival.png', bbox_inches='tight', dpi=150)
plt.show()
print("✅ Plot 2 saved as 'plot2_class_survival.png'")
```

```python
# ── Plot 3: Age Distribution Histogram ───────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Overall age distribution
axes[0].hist(df['age'], bins=20, color='#3498db', edgecolor='black', alpha=0.8)
axes[0].axvline(df['age'].mean(), color='red', linestyle='--',
                linewidth=2, label=f'Mean Age: {df["age"].mean():.1f}')
axes[0].axvline(df['age'].median(), color='orange', linestyle='--',
                linewidth=2, label=f'Median Age: {df["age"].median():.1f}')
axes[0].set_title('Age Distribution of All Passengers', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Age (Years)', fontsize=12)
axes[0].set_ylabel('Number of Passengers', fontsize=12)
axes[0].legend(fontsize=10)

# Right: Age distribution by survival
survived = df[df['survived'] == 1]['age']
not_survived = df[df['survived'] == 0]['age']
axes[1].hist(not_survived, bins=20, alpha=0.7, color='#e74c3c',
             edgecolor='black', label='Did Not Survive')
axes[1].hist(survived, bins=20, alpha=0.7, color='#2ecc71',
             edgecolor='black', label='Survived')
axes[1].set_title('Age Distribution by Survival Status', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Age (Years)', fontsize=12)
axes[1].set_ylabel('Number of Passengers', fontsize=12)
axes[1].legend(fontsize=10)

plt.tight_layout()
plt.suptitle('Figure 3: Passenger Age Distribution', y=1.02, fontsize=15, fontweight='bold')
plt.savefig('plot3_age_histogram.png', bbox_inches='tight', dpi=150)
plt.show()
print("✅ Plot 3 saved as 'plot3_age_histogram.png'")
```

```python
# ── Plot 4 (Bonus): Survival Rate by Age Group ────────────────────────────────

age_rate = df.groupby('age_group', observed=True)['survived'].mean() * 100

plt.figure(figsize=(10, 5))
bar_colors = ['#1abc9c', '#3498db', '#9b59b6', '#e67e22', '#e74c3c']
bars = plt.bar(age_rate.index.astype(str), age_rate.values,
               color=bar_colors, edgecolor='black', width=0.6)

# Add percentage labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.8,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title('Figure 4: Survival Rate (%) by Age Group', fontsize=15, fontweight='bold')
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Survival Rate (%)', fontsize=12)
plt.ylim(0, 100)
plt.xticks(rotation=15, fontsize=10)
plt.tight_layout()
plt.savefig('plot4_age_group_survival.png', bbox_inches='tight', dpi=150)
plt.show()
print("✅ Plot 4 saved as 'plot4_age_group_survival.png'")
```
